# 1 TimeSeries Index  
* pd.to_datetime()
* pd.date_range(start='', end='', peridos='', freq='') -> timeSeries index를 만들 때 주로 사용
* 연, 월, 일 추출하기

```python
import pandas as pd
import numpy as np
df = pd.read_csv('10Y Rate.csv', index_col=0)
df
```

```python
# pd.to_datetime()

df.index = pd.to_datetime(df.index) # str타입이었던 데이터를 datetime형으로 바꿔주는 함수
df.index
```

```python
# pd.date_range()

timeSeries_index = pd.date_range(start='2000-10-26', end= None, periods=100) # 00.10.26부터 일정한 간격으로 100개의 날짜 생성
timeSeries_index
```

```python
# 연, 월, 일 구하기
print(df.index.year)
print(df.index.month)
print(df.index.day)

df['Year'] = df.index.year
df['Month'] = df.index.month
df['Day'] = df.index.day

df # 각 부분을 컬럼으로 만들어서 df에 넣어주면 보기 편함
```

```python
df['Condition'] = np.where(df['Month'] == df['Month'].shift(1), 0, 1)
df['IndexNumber'] = range(df.shape[0])
index_number = df[df['Condition'] == 1]['IndexNumber'].values - 1
df_condition = df.iloc[index_number[1:]]
df_monthly_return = df_condition['Close'].pct_change()
df_monthly_return
```

# 2 Format
* 내가 불러온 데이터 형식이 'str'로 되어있다면?

```python
df = pd.read_csv('data.csv', encoding= 'cp949')
df.info() # Dtype이 object인 것은 여러 자료형이 섞여있다는 뜻
```

```python
# 종가 데이터의 str형(100,000)에서 ,를 지워주기

df['종가'] = [int(value.replace(',', '')) for value in df['종가']]
df.info() # 종가의 데이터타입이 숫자로 변경된 것을 볼 수 있음
```

```python
df = pd.read_csv('data.csv', encoding='cp949', thousands=',') # thousands=''에 ,를 담으면 파이썬이 알아서 처리해줌
df.info()
```
