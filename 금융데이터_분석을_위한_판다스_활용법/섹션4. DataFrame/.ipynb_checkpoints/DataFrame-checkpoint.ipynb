## 2.1 DataFrame 만들기

pd.DataFrame(data, index, columns)  
* index와 columns는 변수를 리스트로 받음
* dict를 data로 받을 때는 index를 지정해 주어야함(원래 지정되어 있지 않아 에러가 발생)

```python
# df 생성하기
import pandas as pd

ls = ['a', 'b', 'c', 'd', 'e'] # 리스트 생성

ls_df = pd.DataFrame(ls, index=['1', '2', '3', '4', '5'], columns=["알파벳"]) # df로 만들어 줌
ls_df
```

```python
# df 생성하기2

ls_df = pd.DataFrame(ls) # df로 만들어 줌
ls_df.index = ['1', '2', '3', '4', '5'] # index 설정
ls_df.columns=["알파벳"] # col 설정

ls_df
```

```python
# dict를 사용해 df를 만드는 법1
dic = {"영수": 101, "영철": 102, "영삼": 103, "영자": 104, "영호": 105}

dic_df = pd.DataFrame(dic, index=[0]) # 이 때 반드시 index를 지정해주어야 함
dic_df
```

```python
# dict를 사용해 df를 만드는 법1

dic_df2 = pd.DataFrame(dic.values()) # 리스트와 유사한 형태로 우선 생성
print(dic_df2)

dic_df2.index = dic.keys() # index를 dict의 key값으로 넣어주기
dic_df2.columns = ['성적']
print(dic_df2)
```

## 2.8 About Nan Value
* df.isna(): NaN값이라면 True를 반환
* df.dropna(): NaN값이 있는 행을 삭제
* df.fillna(value): NaN값을 value로 채워넣음

```python
import pandas as pd
pd.set_option('display.max_rows', None)

df_samsung = pd.read_excel('samsung_naver_excel.xlsx', sheet_name='samsung', index_col=0)
df_naver = pd.read_excel('samsung_naver_excel.xlsx', sheet_name='naver', index_col=0)
lotte = pd.read_csv('롯데케미칼.csv', index_col=0, header=0)
lotte.index = pd.to_datetime(lotte.index)
df = pd.concat([df_naver, df_samsung, lotte], keys=['naver', 'samsung', 'lotte'], names=['stock', '날짜'], axis=1, join = 'outer')
df
```

```python
df_samsung_close = df[[('samsung', 'Close')]]
df_samsung_close
```

```python
df_samsung_close.isna().sum() # Na인 값을 찾으면 True로 출력 -> 따라서 sum()의 결과값은 Na의 값
```

```python
df_samsung_close.dropna() # Na를 삭제
```

```python
df_samsung_close.fillna(0) # Na를 0으로 채워줌
```

```python
# 값이 NaN인 날짜를 구해보자
df_samsung_close[df_samsung_close.isna().values].index # df.isna()의 리턴값은 df이기 때문에 bool타입인 values만 떼서 슬라이싱해주어야 함
```

## 2.10 DataFrame간의 연산
* 같은 index이면서 같은 Column값을 가진 두 value는 연산이 가능
* 두 df의 값 중 하나라도 NaN이 있다면 NaN으로 채워짐
* index나 column이 둘 중 하나에만 존재한다면 NaN으로 채워짐  

* DataFrame[Series의 조건값] -> df를 슬라이싱할 때는 Series의 조건값을 넣어줘야 오류가 나지 않음

```python
# df1 생성
import pandas as pd
import numpy as np

df1 = pd.DataFrame({'a': [1, 4, 7], 'b': [2, 5, 8], 'c': [3, 6, 9]})
df1
```

```python
# df2 생성
df2 = pd.DataFrame({'a': [-1, -4, -7], 'b': [np.NaN, np.NaN, np.NaN], 'e': [3, 6, 9]})
df2
```

```python
# 두 df 합쳐보기
df1 + df2 # 둘 모두 값이 존재하는 a는 더해져서 0, NaN값이 존재하는 B는 NaN, 1에만 있는 c, 2에만있는 e는 NaN으로 채워짐
```

```python
pd.set_option('display.max_rows', None)

df_samsung = pd.read_excel('samsung_kospi_excel.xlsx', sheet_name='samsung', index_col=0)
df_kospi = pd.read_excel('samsung_kospi_excel.xlsx', sheet_name='kospi', index_col=0)
```

## 2.11 Sorting
index에 대한 sort
* df.sort_index(ascending = True or False)  

Value에 대한 sort
* df.sort_values(by = '컬럼이름', ascending = True or False)

```python
pd.set_option('display.max_rows', None)

df_samsung = pd.read_excel('samsung_kospi_excel.xlsx', sheet_name='samsung', index_col=0)
df_kospi = pd.read_excel('samsung_kospi_excel.xlsx', sheet_name='kospi', index_col=0)
```

```python
# 필요한 값들만 뽑아 새로운 df생성
df = pd.DataFrame()

df['Kospi_Return'] = df_kospi['Change']
df['samsung_Return'] = df_samsung['Change']
```

```python
# 코스피가 2%이상 상승하고, 삼성전자가 2%이상 상승한 날짜는 언제인가?

df['Condition'] = np.where(((df['Kospi_Return'] >= 0.02) & (df['samsung_Return'] >= 0.02)), True, np.NaN) # 두 조건 시리즈를 넣어주고 조건에 맞으면 True, 아니면 NaN을 넣어줌

df_result = df.dropna() # na값을 날려주면 조건에 부합한 날짜만 나옴

df_result
```

```python
# index sorting

df_result.sort_index(ascending=True) # 날짜 오름차순
```

```python
# value sorting
df_result.sort_values(by = 'samsung_Return', ascending=False) # 삼성전자 상승률이 높은 순
```

## 2.14 GroupBy
df.groupby(['컬럼이름']).함수()  
* 그룹핑을 한 뒤, 함수를 하나 더 사용해 그룹별 통계를 낼 수 있음

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)

df = pd.read_excel('groupby_example.xlsx')
df # 이상한 것들이 섞여나온 df라는 사실을 알 수 있음
```

```python
# header을 바꿔서 일단 col은 정상화 성공
df = pd.read_excel('groupby_example.xlsx', header=6)
df
```

```python
# NaN값을 가진 열들 전부 drop
df = df.dropna(axis=1)
df
```

```python
df.columns = ['기업', 'Sector', 'PER'] # 컬럼명 지정
df.index = df['기업'] # df의 index를 기업(원래 인덱스)로 지정
df
```

```python
df = df.drop('기업', axis=1) # 기업 컬럼 삭제
df
```

```python
# groupby()를 이용해 각 섹터별 PER이 평균보다 낮은 회사는?
underavg_che = df.groupby(['Sector']).mean().loc['화학'].values[0] # 화학 섹터의 평균 PER구하기
underavg_che
```

```python
df_che = df[df['Sector'] == '화학'].copy() # 섹터가 화학인 회사들을 추출
# 위에서와 같이 df를 필터링하면 복사본인지 원본데이터인지 명확하지 않은 상황 발생 -> copy()를 사용해 확실히 복사본으로 만들어줌
df_che['Condition'] = np.where(df_che['PER'] <= underavg_che, True, np.NaN)
df_che.dropna() # 평균 이상인 것들 제거
```
