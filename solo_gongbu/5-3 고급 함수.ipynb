```python
# 람다: 매개변수로 함수를 사용하는 함수(함수포인터 같은 개념인듯)

def call_10_times(func):
    for i in range(10):
        func()

def hello():
    print("ㅎㅇ")

call_10_times(hello)
```

```python
# map함수: map(함수, 리스트) -> 리스트의 요소를 함수에 넣어서 리턴된 값을 리스트 형식으로 반환
# filter 함수: filter(함수, 리스트) -> 리스트의 요소를 함수에 넣은 뒤 True인 것을 리스트 형식으로 반환

def power(item):
    return item * item

def under_3(item):
    return item < 3


list_input_a = [1, 2, 3, 4, 5]

# map() 함수 사용
output_a = map(power, list_input_a)
print("map 함수 실행결과: ", output_a) # 포인터로 반환하는 듯
print("map 함수 실행결과: ", list(output_a))

print()
# filter() 함수 사용
output_b = filter(under_3, list_input_a)
print("filter 함수 실행결과: ", output_b) # 포인터로 반환하는 듯
print("filter 함수 실행결과: ", list(output_b)) # 포인터로 반환하는 듯
```

```python
# 위처럼 함수 구문을 따로 작성하는게 귀찮다고 생각함 -> 그래서 나온게 람다
# lambda 매개변수: 리턴값

# 함수 선언
power = lambda x: x * x
under_3 = lambda x: x < 3

# map() 함수 사용
output_a = map(power, list_input_a)
print("map 함수 실행결과: ", output_a) # 포인터로 반환하는 듯
print("map 함수 실행결과: ", list(output_a))

print()
# filter() 함수 사용
output_b = filter(under_3, list_input_a)
print("filter 함수 실행결과: ", output_b) # 포인터로 반환하는 듯
print("filter 함수 실행결과: ", list(output_b)) # 포인터로 반환하는 듯
```

```python
# 인라인 람다
# 람다는 함수포인터 자리에 바로 집어넣을 수 있음

# map() 함수 사용
output_a = map(lambda x: x * x, list_input_a)
print("map 함수 실행결과: ", output_a) # 포인터로 반환하는 듯
print("map 함수 실행결과: ", list(output_a))

print()
# filter() 함수 사용
output_b = filter(lambda x: x < 3, list_input_a)
print("filter 함수 실행결과: ", output_b) # 포인터로 반환하는 듯
print("filter 함수 실행결과: ", list(output_b)) # 포인터로 반환하는 듯
```

```python
# with 키워드 이용해 파일 열기
with open("basic.txt", "w") as file:
    # 파일 쓰기
    file.write("Hello Python Programming..!")

# 스코프가 끝나면 자동으로 닫힘
```

```python
# 텍스트 한 줄씩 읽기

# CSV파일 같은 경우 헤더와 데이터로 이루어져 있는데
# 한 줄에 하나의 객체가 담기므로 한 줄씩 잘라 읽는 것이 중요힘
```

```python
# 랜덤하게 1000명의 이름, 키, 몸무게를 담고 있는 파일 만들기

# 랜덤 모듈
import random

# 간단한 한글 리스트 생성
hangul = list("가나다라마바사아자차카타파하")

# 파일을 쓰기 모드로 열기
with open("info.txt", "w") as file:
    for i in range(1000):
        # 랜덤한 값으료 변수 생성
        name = random.choice(hangul) + random.choice(hangul)
        weight = random.randrange(40, 100)
        height = random.randrange(140, 200)
        
        # 텍스트 작성
        file.write("{}, {}, {}\n".format(name, weight, height))
```

```python
# 한 줄씩 처리하기
# 한 줄씩 처리하려면 'for 한 줄 in 파일:' 형식으로 반복해 처리

with open("info.txt", "r") as file:
    for line in file:
        # 변수 선언
        (name, weight, height) = line.strip().split(', ')

        # 데이터 검사
        if (not name) or (not weight) or (not height):
            continue

        # 결과 계산
        bmi = int(weight) / ((int(height) / 100) ** 2)
        result = ""
        if 25 <= bmi:
            result = '과체중'
        if 18.5 <= bmi:
            result = '정상 체중'
        else:
            result = '저체중'

        # 출력합니다.
        print('\n'.join([
        "이름: {}",
        "몸무게: {}",
        "키: {}",
        "bmi: {}",
        "결과: {}",
        ]).format(name, weight, height, bmi, result))
        print()
```
