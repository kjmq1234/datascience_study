```python
# beautifulsoup 모듈: 웹페이지 분석 모듈
from urllib import request
from bs4 import BeautifulSoup

# urlopen()함수로 기상청의 전국날씨를 읽기
target = request.urlopen("http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108")

# BeautifulSoup를 통해 웹페이지 분석
soup = BeautifulSoup(target, "html.parser") # 클래스를 생성

# location 태그를 찾기
for location in soup.select("location"): # select()함수를 통해 태그를 선택(여러 개 선택 가능) 
    print("도시:", location.select_one("city").string) # select_one()함수를 통해 태그를 선택(하나만 선택 가능)
    print("날씨:", location.select_one("wf").string)
    print("최저기온:", location.select_one("tmn").string)
    print("최고기온:", location.select_one("tmx").string)
    print()
```

```python
# 환경변수를 사용해 로컬호스트에 출력
from flask import Flask
app = Flask(__name__)

@app.route('/') # @부분은 데코레이터
def hello():
    return "<h1>Hello World!</h1?"
```

```python
# 라이브러리와 프레임워크의 차이

# 라이브러리: 정상적인 제어를 하는 모듈
# 프레임워크: 제어의 역전이 발생하는 모듈

# 제어의 역전(IoC) -> 코드 내부를 구현하지 않아 흐름이 사용자의 마음대로 끌려오는 것

# 라이브러리 함수는 구현이 되어있어 함수를 꺼내서 흐름대로 사용하면 되지만
# 프레임워크는 구현이 되어 있지않아 사용자에게 제어권이 넘어간다는 특징이 있음 -> 이게 IoC
```

```python
# 데코레이터: @로 시작하는 구문
# 함수 데코레이터: 함수의 앞뒤에 꾸밀 부가적/반복 내용을 손쉽게 사용할 수 있도록 한 것


# 'hello'를 출력하는 함수
def hello():
    print("hello")

# 이제 데코레이터를 통해 hello의 앞뒤에 '인사가 시작/종료 되었습니다'를 출력하려고 함
```

```python
# 함수 데코레이터를 생성
def test(function):
    def wrapper():
        print("인사가 시작되었습니다.")
        function()
        print("인사가 종료되었습니다.")
    return wrapper


# 데코레이터를 붙여 함수를 만들기
@test
def hello():
    print("hello")


hello()
```
